## Scrape Recipe Metadata (part 1)

### Recipe Name, Star Rating, Number of Ratings, Number of Reviews, Recipe Bleb

```python
# create empty lists to store data
url_ID = []
recipe_name = []
star_rating = []
number_of_rankings = []
number_of_reviews = []
blerb = []
```

```python
for url in good_urls:
    
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'lxml')
    
    url_scrape = url
    url_ID.append(url_scrape)

    try:
        recipe_name_scrape = soup.find("h1", {"class": "headline heading-content"}).text.strip()
        recipe_name.append(recipe_name_scrape)
    except:
        recipe_name.append(np.nan)
    try:
        number_of_rankings_scrape = soup.find("span", {"class": "ugc-ratings-item"}).text.strip()
        number_of_rankings.append(number_of_rankings_scrape)
    except:
        number_of_rankings.append(np.nan)
    try:
        number_of_reviews_scrape = soup.find("a", {"class": "ugc-ratings-link ugc-reviews-link"}).text.strip()
        number_of_reviews.append(number_of_reviews_scrape)
    except:
        number_of_reviews.append(np.nan)
    try:
        blerb_scrape = soup.find("p", {"class": "margin-0-auto"}).text.strip()
        blerb.append(blerb_scrape)
    except:
        blerb.append(np.nan)
    try:
        stars_section = soup.find_all("a", {"class": "add-rating-reviews-link ugc-ratings-link review-stars-container"})
        for section in stars_section:
            stars = section.find_all("span", {"class": "rating-star active"})
        stars_rating_scrape = len(stars)   
        star_rating.append(stars_rating_scrape)   
    except:
        star_rating.append(np.nan)
```

```python
metadata_table = pd.DataFrame()

metadata_table["url"] = url_ID
metadata_table["recipe_name"] = recipe_name
metadata_table["star_rating"] = star_rating
metadata_table["number_of_rankings"] = number_of_rankings
metadata_table["number_of_reviews"] = number_of_reviews
metadata_table["blerb"] = blerb

metadata_table
```

## Scrape Recipe Metadata (part 2)

### Prep Time, Cook Time, Additional Time, Total Time, Servings, Yield

```python
# create empty lists to store data
url_ID2 = []
prep_time = []
cook_time = []
additional_time = []
total_time = []
servings = []
recipe_yield = []
columnheaders = ['prep', 'cook', 'additional', 'total', 'servings', 'yield']


for url in good_urls:
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'lxml')
    
    url_scrape = url
    url_ID2.append(url_scrape)
    metadataitems = soup.find_all("div", {"class":"recipe-meta-item"})
    metadata = {}
    for item in metadataitems:    
        header = item.find("div", {"class":"recipe-meta-item-header"}).text.lower().strip(":")
        body = item.find("div", {"class":"recipe-meta-item-body"}).text
        
        metadata[header] = body
        
    for header in columnheaders:
        if header in metadata.keys():
            data = metadata[header]
            
            if header == "prep":
                prep_time.append(data)
            elif header == "cook":
                cook_time.append(data)
            elif header == "additional":
                additional_time.append(data)
            elif header == "total":
                total_time.append(data)
            elif header == "servings":
                servings.append(data)
            elif header == "yield":
                recipe_yield.append(data)
        else:
            if header == "prep":
                prep_time.append(np.nan)
            elif header == "cook":
                cook_time.append(np.nan)
            elif header == "additional":
                additional_time.append(np.nan)
            elif header == "total":
                total_time.append(np.nan)
            elif header == "servings":
                servings.append(np.nan)
            elif header == "yield":
                recipe_yield.append(np.nan)

        
```

```python
metadata_table2 = pd.DataFrame()

metadata_table2["url"] = url_ID2
metadata_table2["prep_time"] = prep_time
metadata_table2["cook_time"] = cook_time
metadata_table2["additional_time"] = additional_time
metadata_table2["total_time"] = total_time
metadata_table2["servings"] = servings
metadata_table2["recipe_yield"] = recipe_yield

metadata_table2
```

## Scrape Recipe Directions

### Step Number, Directions

```python
# create empty lists to store data
steps = []
directions = []
url3 = []

for url in good_urls:
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'lxml')

    items = soup.find_all("li", {"class":"subcontainer instructions-section-item"})
    for item in items:    
        url_scrape = url
        url3.append(url_scrape)
        step = item.find("span", {"class": "checkbox-list-text"}).text
        p = item.find("div", {"class":"paragraph"}).text

        steps.append(step)
        directions.append(p)
```

```python
directions_table = pd.DataFrame()

directions_table["url"] = url3
directions_table["steps"] = steps
directions_table["directions"] = directions

directions_table
```

## Scrape Ingredients

```python
# create empty lists to store data
url_ID3 = []
ingredient_list = []

for url in good_urls:
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'lxml')
    
    ingredients = soup.find_all("span", {"class": "ingredients-item-name"})
    for ingredient in ingredients:
        url_scrape = url
        url_ID3.append(url_scrape)
        item = ingredient.text.strip()
        ingredient_list.append(item)
```

```python
ingredient_table = pd.DataFrame()

ingredient_table["url"] = url_ID3
ingredient_table["ingredient_list"] = ingredient_list

ingredient_table.head(50)
```

# TRANSFORM

## Split full ingredient string into Quantity, Unit of Measure, and Ingredient using delimiters

```python
import re
quant = []
unit_of_measure = []
ingredient = []

regexp = re.compile(r'\(.\d|\(\d|fluid ounces|fluid ounce|ounces|ounce|pkg.|pound|drops|drop| cans | can |containers|container| cups | cup |teaspoons|teaspoon|tiny pinch|pinch|tablespoons|tablespoon')

for x in ingredient_table['ingredient_list']:
    
    if x[0].isnumeric():
        if regexp.search(x):
            string_list = re.split('(\(.\d|\(\d|fluid ounces|fluid ounce|ounces|ounce|pkg.|pound|drops|drop| cans | can |container| cups | cup |teaspoons|teaspoon|tiny pinch|pinch|tablespoons|tablespoon)', x, maxsplit=1)
            quant.append(string_list[0])
            
            measure_temp = string_list[1]
            ingred_temp = string_list[2]
            if "(" in measure_temp:
                ingred_temp = measure_temp + ingred_temp
                measure_temp = "unit(s)"
                            
            unit_of_measure.append(measure_temp)
            ingredient.append(ingred_temp)
            
        else:
            string_list = re.split(" ",x,1)
            quant.append(string_list[0])
            unit_of_measure.append("unit(s)")
            ingredient.append(string_list[1])
    
            
    else:
        quant.append(np.nan)
        unit_of_measure.append("unit(s)")
        ingredient.append(x)
            

```

```python
quantdf=pd.DataFrame(quant)
quantdf.value_counts()
```
